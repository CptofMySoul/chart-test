```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipyfilechooser import FileChooser
from IPython.display import display
import os
import ipywidgets as widgets

plt.style.use('/Users/rumcglot/plotStyle.mplstyle') # set up plot styling
set_matplotlib_formats('svg')

# Create new FileChooser:
# Path: current directory
# File: data.txt
# Title: <b>Select Data File</b>
# Show hidden files: no
# Use the default path and filename as selection: yes
# Use folder icons: yes
fdialog = FileChooser(
    os.getcwd(),
    filename='data.csv',
    title='<b>Select Data File</b>',
    show_hidden=False,
    select_default=True,
    use_dir_icons=True
)
```

```python

x_axis_title=widgets.Text(
    value='Safety Current Limit',
    placeholder='Enter the x-axis title:',
    description='X Axis:',
    disabled=False
)

display(x_axis_title)
```

```python
x_axis_title.value
```

## Select the data file

```python
display(fdialog)
```

## Import .csv with two columns of data â€“ x, y

```python
xy_data = pd.read_csv(fdialog.selected, names=['x','y'])
```

## Enter in specific chart info

```python
y_axis_label = "Safety Limiting Current (mA)"
x_axis_label = "Ambient Temperature (ËšC)"

y_axis_min_max = (0, 100)
x_axis_min_max = (0, 160)

chart_text = "VDDA, VDDB = 24 V"
```

```python
fig, ax = plt.subplots(1)

# Call plot() method on the appropriate object
ax.plot(xy_data[['x']], xy_data[['y']])

ax.set_ylabel(y_axis_label)
ax.set_xlabel(x_axis_label)
ax.set_ylim(y_axis_min_max[0], y_axis_min_max[1])
ax.set_xlim(x_axis_min_max[0], x_axis_min_max[1])

ax.text(0.75, 0.8,chart_text,
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes,  bbox={'facecolor': 'white', 'edgecolor': 'none'})
```
